```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

## Applied TSNE

```python
mnist = fetch_openml('mnist_784', as_frame=False)

X = mnist.data[:5000]
y = mnist.target[:5000]
y = y.astype(int) 
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.colorbar()
plt.title('t-SNE Visualization of MNIST')
plt.show()
```

## Applied PCA

```python
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.colorbar()
plt.title('PCA Visualization of MNIST')
plt.show()
```

## Applied LDA(Linear Discriminant Analysis)

```python
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X,y)
X_lda = lda.transform(X)
```

```python
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y)
plt.colorbar()
plt.title('LDA Visualization of MNIST')
plt.show()
```

## Applied GaussianRandomProjection 

```python
from sklearn.random_projection import GaussianRandomProjection

grp = GaussianRandomProjection(n_components=2)
X_grp = grp.fit_transform(X)
```

```python
plt.scatter(X_grp[:, 0], X_grp[:, 1], c=y)
plt.colorbar()
plt.title('Gaussian Random Projection Visualization of MNIST')
plt.show()
```
